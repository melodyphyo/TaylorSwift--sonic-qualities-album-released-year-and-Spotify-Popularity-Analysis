```python
track13 = df[df['track_number'] == 13].groupby('name')['popularity'].max()
```

```python
plt.plot(track13.index, track13.values, color = 'orange', label = 'Track13', marker = 'o')

plt.xticks(rotation = 90)
plt.legend()
plt.xlabel('Song Name')
plt.ylabel('Popularity')
plt.title('Popularity of Track number 13 by Taylor Swift')
plt.show()
```

```python
album_pop = df.groupby('release_year')['popularity'].max()
custom_palette = sns.color_palette('husl', len(album_pop))
sns.barplot(x=album_pop.index, y=album_pop.values, palette = custom_palette)
plt.xticks(rotation = 90)
plt.title('Popularity of Taylor Swift\'s Career Upon the Album Released Year')
plt.xlabel('Album Release Year')
plt.ylabel('Popularity')
plt.show()
```

```python
most_popular_songs = df.loc[df.groupby('album')['popularity'].idxmax()][['name', 'album', 'popularity']]
```

```python
#most_popular_songs = df.groupby(['name', 'album'])['popularity'].idxmax()

most_popular_songs = df.loc[df.groupby('album')['popularity'].idxmax()][['name', 'album', 'popularity']].sort_values(by = ['popularity'], ascending = False)
```

```python
most_popular_songs
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(
    data=most_popular_songs,
    x='popularity',
    y='name',
    color = "orange"
)
plt.title('Most Popular Song Per Album', fontsize=14)
plt.xlabel('Popularity')
plt.ylabel('Song name from Each Album')
plt.tight_layout()
plt.show()
```

```python
# filter the row  which include the "voice memo" in the song name.

df = df[~df['name'].str.contains("voice memo", case=False, na=False)]
```

```python
least_popular_songs = df.loc[df.groupby('album')['popularity'].idxmin()][['name', 'album', 'popularity']].sort_values(by = ['popularity'], ascending = False)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(
    data=least_popular_songs,
    x='popularity',
    y='name',
    color = "orange"
)
plt.title('Least Popular Song Per Album', fontsize=14)
plt.xlabel('Popularity')
plt.ylabel('Song name from Each Album')
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df,x='popularity', y='acousticness', hue = 'album')
sns.regplot(data=df,x='popularity', y='acousticness', scatter= False, color = "black")
plt.legend(title = "Album", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
